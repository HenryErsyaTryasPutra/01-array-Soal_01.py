```python


import matplotlib.pyplot as plt

def input_nilai(n=10):
    nilai = []
    print(f"Masukkan {n} nilai mahasiswa:")
    for i in range(n):
        while True:
            try:
                x = float(input(f"Nilai ke-{i+1}: "))
                nilai.append(x)
                break
            except ValueError:
                print("Input harus berupa angka!")
    return nilai

def hitung_statistik(nilai):
    tertinggi = max(nilai)
    terendah = min(nilai)
    rata2 = sum(nilai) / len(nilai)
    lulus = sum(1 for x in nilai if x >= 60)
    tidak_lulus = len(nilai) - lulus

    return tertinggi, terendah, rata2, lulus, tidak_lulus

def grafik_tertinggi_terendah(nilai):
    tertinggi = max(nilai)
    terendah = min(nilai)

    kategori = ["Tertinggi", "Terendah"]
    data = [tertinggi, terendah]

    plt.figure()
    plt.bar(kategori, data)
    plt.title("Grafik Nilai Tertinggi dan Terendah")
    plt.xlabel("Kategori")
    plt.ylabel("Nilai")
    plt.show()

def grafik_kelulusan(nilai):
    lulus = sum(1 for x in nilai if x >= 60)
    tidak_lulus = len(nilai) - lulus

    kategori = ["Lulus", "Tidak Lulus"]
    data = [lulus, tidak_lulus]

    plt.figure()
    plt.bar(kategori, data)
    plt.title("Grafik Kelulusan Mahasiswa")
    plt.xlabel("Kategori")
    plt.ylabel("Jumlah")
    plt.show()


def main():
    nilai = input_nilai(10)
    tertinggi, terendah, rata2, lulus, tidak_lulus = hitung_statistik(nilai)

    print("\n=== HASIL ===")
    print("Nilai:", nilai)
    print("Nilai Tertinggi:", tertinggi)
    print("Nilai Terendah:", terendah)
    print("Rata-rata:", round(rata2, 2))
    print("Jumlah Lulus (>=60):", lulus)
    print("Jumlah Tidak Lulus:", tidak_lulus)

    grafik_tertinggi_terendah(nilai)
    grafik_kelulusan(nilai)

if __name__ == "__main__":
    main()
```
